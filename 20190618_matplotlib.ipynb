```python
import pandas 
import numpy as np
```

```python
df1 = pandas.read_csv('dow_1year_ago.csv')
df2 = pandas.read_csv('dow_2year_ago.csv')
arr = np.concatenate((df2,df1), axis=0)
arr
```

```python
#Date,Open,High,Low,Close,Adj Close,Volume
plt.figure(figsize=(40,10)) #width, hegiht in inch
plt.title('Dow Jones')
plt.plot(arr[:, 1], label='Open', linestyle='--')
plt.plot(arr[:, 2], label='High', linestyle=':')
plt.plot(arr[:, 3], 'r.', label='Low', )
plt.plot(arr[:, 4], label='Close')
plt.legend(loc=4)

```

```python
#Date,Open,High,Low,Close,Adj Close,Volume
plt.figure(figsize=(40,10)) #width, hegiht in inch
plt.title('Dow Jones')
plt.plot(arr[:, -1], label='Volume', linestyle='--')
plt.legend(loc=4)
```
