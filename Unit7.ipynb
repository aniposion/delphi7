```python
import csv
import matplotlib.pyplot as plt


f = open('C:/Users/wtime/Desktop/Python/unit7.csv','r')
data = csv.reader(f)
header = next(data)
#print(header)

result=[]
search=input('검색하고 싶은 동을 입력하세요.')
for row in data:     
    if search in row[0]:
        for i in row[3:]:
            #print(i)
            result.append(int(i))

f.close()
```

```python

plt.style.use('ggplot')
plt.plot(result)
plt.show()
```
